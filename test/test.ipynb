```python
import serpapi
import os
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv('SERPAPI_KEY')
client = serpapi.Client(api_key=api_key)
```

```python
query = input("enter a input:")
```

```python
result = client.search(
    q=query,
    engine="google_scholar",
    hl="en",
    gl="us"
).as_dict()
```
